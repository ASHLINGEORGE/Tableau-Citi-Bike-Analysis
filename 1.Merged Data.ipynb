```python
# Dependencies
import pandas as pd
```

```python
# List of file names to be combined 
files18 = ['202201.csv', '202202.csv', '202203.csv', '202204.csv', '202205.csv', '202206.csv', '202207.csv', '202208.csv', '202209.csv', '202210.csv', '202211.csv', '202212.csv']

# Create an empty list to store each dataframe
df18_list = []

# Loop through each file in the list
for file in files18:
    # Read the file into a dataframe
    df18 = pd.read_csv(file)

    # Append the dataframe to the list
    df18_list.append(df18)

# Concatenate all dataframes in the list into one dataframe
df18 = pd.concat(df18_list)

# Show dataframe
df18
```

```python
# List of file names to be combined 
files19 = ['202301.csv', '202302.csv', '202303.csv', '202304.csv', '202305.csv', '202306.csv', '202307.csv', '202308.csv', '202309.csv', '202310.csv', '202311.csv', '202312.csv']

# Create an empty list to store each dataframe
df19_list = []

# Loop through each file in the list
for file in files19:
    # Read the file into a dataframe
    df19 = pd.read_csv(file)

    # Append the dataframe to the list
    df19_list.append(df19)

# Concatenate all dataframes in the list into one dataframe
df19 = pd.concat(df19_list)

# Show dataframe
df19
```

```python
# Concatenate all dataframes for all the years into one
combined_df = pd.concat([df18, df19], axis=0)

# Display the results
combined_df
```

```python
combined_df = combined_df.rename(columns={
    'started_at': 'started_time',
    'ended_at': 'ended_time',
    'start_lat': 'start_latitude',
    'end_lat': 'end_latitude',
    'start_lng': 'start_longitude',
    'end_lng': 'end_longitude'
})
```

```python
# Save the dataframe to a CSV file
combined_df.to_csv('citibike.csv', index=False)
```
